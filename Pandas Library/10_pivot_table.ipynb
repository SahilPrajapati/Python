```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Dell\Documents\GitHub\py\pandas\10_pivot\weather.csv')
df
```

```python
#creating pivot having rows of date and columns of city
df.pivot(index = 'date',columns='city')
```

```python
#creating pivot having rows of date and columns of city of particular type eg humidity
df.pivot(index = 'date',columns='city',values = 'humidity')
```

```python
#pivot table of default aggregate function mean.
print (df)

df.pivot_table(index = 'date',columns = 'city',aggfunc = 'sum')
```

```python
df2 = pd.read_csv(r'C:\Users\Dell\Documents\GitHub\py\pandas\10_pivot\weather2.csv')
df2
```

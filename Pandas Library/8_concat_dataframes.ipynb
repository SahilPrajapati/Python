```python
import pandas as pd

df1 = pd.DataFrame({
    'city':['mumbai','delhi','bangalore'],
    'temperature':[32,45,30],
    'humidity':[80,60,78]
})

df1
```

```python
df2 = pd.DataFrame({
    'city':['new york','chicago','oralando'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})

df2
```

```python
#concating...
df = pd.concat([df1,df2])
df
```

```python
#concatenating in proper index
df = pd.concat([df1,df2],ignore_index = True)
df
```

```python
#associating keys with each dataframe
df = pd.concat([df1,df2],keys=['Data Frame 1','Data Frame 2'])
df
```

```python
#printing a particular data frame with associated key

df.loc['Data Frame 2']
```

```python
#concatenating column wise...
df3= pd.DataFrame({
    'city':['new york','chicago','oralando'],
    'temperature':[21,14,35],
})

print(df3,'\n')
df4 = pd.DataFrame({
    'city':['new york','chicago','oralando'],
    'humidity':[68,65,75]
})
print(df4,'\n')


d = pd.concat([df3,df4],axis = 'rows',sort=False)
print (d)
```

```python
#concating with series

s = pd.Series(['humid','dry','rain'],name = 'event')
df5 = pd.concat([df3,s],axis = 'columns')
df5
```

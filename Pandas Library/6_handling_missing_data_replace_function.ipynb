```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\Dell\Documents\GitHub\py\pandas\6_handling_missing_data_replace\weather_data.csv')
df
```

```python
new_df = df.replace([-88888,-99999],[6,3])
new_df
```

```python
new_df = df.replace({
    'temperature':-99999,
    'windspeed':-88888,
    'event':'0'
},np.NaN)

new_df
```

```python
new_df = df.replace({
    -99999:0,
    -88888:'hello',
    '0':'no event'
})

new_df
```

```python
#replace by using regular expressions(regex)

new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex = True)

new_df
```
